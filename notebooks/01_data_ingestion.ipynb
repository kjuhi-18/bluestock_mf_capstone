### AMFI Scheme Code Structure

AMFI (Association of Mutual Funds in India) assigns a unique numeric
scheme code to each mutual fund scheme.

The AMFI code serves as:
- Unique scheme identifier
- Primary key for linking datasets
- Reference used by NAV providers and APIs
- Join key between fund master and NAV history

Example:
125497 → HDFC Top 100 Fund Direct Growth
119551 → SBI Bluechip Fund

Validate AMFI Codes

```python
fund_codes = set(
    fund_master['amfi_code']
)

nav_codes = set(
    nav_history['amfi_code']
)

matching_codes = fund_codes.intersection(nav_codes)

missing_codes = fund_codes - nav_codes

extra_codes = nav_codes - fund_codes
```

Validation Summary

```python
print("="*50)
print("AMFI CODE VALIDATION")
print("="*50)

print(f"Fund Master Codes : {len(fund_codes)}")
print(f"NAV History Codes : {len(nav_codes)}")
print(f"Matching Codes    : {len(matching_codes)}")
print(f"Missing Codes     : {len(missing_codes)}")
print(f"Extra Codes       : {len(extra_codes)}")

coverage = (
    len(matching_codes)
    / len(fund_codes)
) * 100

print(f"\nCoverage: {coverage:.2f}%")
```

Display Missing Codes (if any)

```python
if len(missing_codes) > 0:
    print("\nMissing Codes:")
    print(list(missing_codes)[:20])
else:
    print("\nAll AMFI codes are present in NAV history.")
```
